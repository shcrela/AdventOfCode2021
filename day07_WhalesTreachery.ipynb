```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initial_positions = [16,1,2,0,4,2,7,1,2,14]
initial_positions = np.loadtxt("data/day07a.txt", delimiter=",", dtype=int)
M = max(initial_positions)
N = len(initial_positions)
iarr = np.array(initial_positions).reshape(1, N) # a line

# each line is a N times repeated position from 0 to M (included)
positions = np.arange(M+1).reshape(M+1, 1) @ np.ones((1, N), dtype=int) # shape M x N

# How much fuel would everybody spend if they were to go to the given position
fuel_consumption = np.abs(iarr - positions) # N crabs x M positions

total_fuel = np.sum(fuel_consumption, axis=1)

plt.plot(total_fuel);
```

```python
# The result:
np.min(total_fuel)
```

```python
# initial_positions = [16,1,2,0,4,2,7,1,2,14]
initial_positions = np.loadtxt("data/day07a.txt", delimiter=",", dtype=int)
M = max(initial_positions)
N = len(initial_positions)
iarr = np.array(initial_positions).reshape(1, N) # a line

# each line is a N times repeated position from 0 to M (included)
positions = np.arange(M+1).reshape(M+1, 1) @ np.ones((1, N), dtype=int) # shape M x N

# How much fuel would everybody spend if they were to go to the given position
n_paces = np.abs(iarr - positions) # N crabs x M positions
# New way of calculating fuel consumption:
fuel_consumption = n_paces*(n_paces + 1) / 2
total_fuel = np.sum(fuel_consumption, axis=1)

plt.plot(total_fuel);
```

```python
# The result:
int(np.min(total_fuel))
```
