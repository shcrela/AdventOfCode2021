```python
import numpy as np

filename = "data/day04a.txt"
sequence = np.loadtxt(filename, max_rows=1, delimiter=",", dtype=int)

board_lines = []
i = -2
with open(filename, "r") as ff:
    for line in ff:
        if len(line) == 15:
            board_lines.append([int(s) for s in line.split()])

boards = np.array(board_lines).reshape(-1, 5, 5)
```

```python
marked_numbers = np.zeros_like(boards, dtype=bool)
unfilled_boards = np.ones(len(boards), dtype=bool)
winning_boards = []
n_board = "none yet"


for broj in sequence:
    new_winner = False
    marked_numbers[boards==broj] = True
    column_filled = np.sum(marked_numbers, axis=1)==5
    line_filled = np.sum(marked_numbers, axis=2)==5
    if np.any(line_filled[unfilled_boards]):
        n_board = list(set(np.where(line_filled)[0]).intersection(set(np.where(unfilled_boards)[0])))
        unfilled_boards[n_board] = False
        winning_boards.extend(n_board)
        new_winner = True
    elif np.any(column_filled[unfilled_boards]):
        n_board = list(set(np.where(column_filled)[0]).intersection(set(np.where(unfilled_boards)[0])))
        unfilled_boards[n_board] = False
        winning_boards.extend(n_board)
        new_winner = True
    print(f"Number: {broj:2d}")
    if new_winner:
        print(f"Congrats to board{'s'*(len(n_board)>1)} N°{n_board} !\n"
              f"We now have {len(set(winning_boards))} winning board{'s'*(len(set(winning_boards))>1)} in total!\n")
    if np.sum(unfilled_boards) == 0:
        break
```

```python
np.sum(boards[n_board][np.logical_not(marked_numbers[n_board])])*broj
```
